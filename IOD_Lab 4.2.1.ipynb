### 2. Set Target Variable

Create a target variable for wine quality.

```python
# Target Variable
y = df['quality']
y
```

### 3. Set Predictor Variables

Create a predictor matrix with variables of your choice. State your reason.

```python
# ANSWER
X = df[['alcohol', 'density', 'volatile acidity', 'chlorides']]
X
```

### 4. Using Linear Regression Create a Model and Test Score

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
# Train-Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
# Create a model for Linear Regression
lm = LinearRegression()
# Fit the model with the Training data
lm.fit(X_train, y_train)
# Calculate the score (R^2 for Regression) for Training Data
print("model score of trained set:", lm.score(X_train, y_train))
# Calculate the score (R^2 for Regression) for Testing Data
print("model score of test set:", lm.score(X_test, y_test))
```

### BONUS: Cross validation

```python
# Cross validation 
from sklearn.model_selection import KFold 
from sklearn.metrics import mean_squared_error
```

```python
print(type(y))
```

```python
# Set up 5-fold cross validation  
k_fold = KFold(5, shuffle=True)
train_scores = []
train_rmse = []
test_scores = []
test_rmse = []

for k, (train, test) in enumerate(k_fold.split(X)):
    # Print and show that the index of predictor matrix is returned
    print('k:', k, 'TRAIN:', train, 'TEST:', test)
    # Get training and test sets for X and y
    X_train, X_test = X.iloc[train,:], X.iloc[test,:]
    # Fit model with training set
    y_train, y_test = y[train], y[test]
    # Make predictions with training and test set
    yhat_train = lm.predict(X_train)
    yhat_test = lm.predict(X_test)
    # Score R2 and RMSE on training and test sets and store in list
    train_scores.append(lm.score(X_train, y_train))
    train_rmse.append(np.sqrt(mean_squared_error(y_train, yhat_train)))
    test_scores.append(lm.score(X_test, y_test))
    test_rmse.append(np.sqrt(mean_squared_error(y_test, yhat_test)))
    
# Create a metrics_df dataframe to display r2 and rmse scores
pd.DataFrame(list(zip(train_scores,train_rmse,test_scores,test_rmse)), columns=['train_scores','train_rmse','test_scores','test_rmse'])
```

```python
# Describe the metrics
# Cross validation returns consistent scores of around 0.25 to 0.27. Our model is highly accurate.
```
